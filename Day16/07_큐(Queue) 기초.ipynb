## 07-1 큐의 개념

- 줄을 서다 라는 의미
- 선입선출 , FIFO 구조
- 삽입연산 : push, 꺼내는 연산 : pop
- 여러 이벤트가 발생했을때, 발생한 순서대로 처리할 때 큐가 사용

### 큐의 특성 활용분야
- 작업 대기열 : 네트워크 통신을 할 대 다수의 클라이언트에서 서버에 작업을 요청하면 서버는 요청이 들어온 순서대로
작업을 처리
- 이벤트 처리 : 어떤 애플리케이션이나 시스템에서 사용자의 이벤트, 예를 들어 키보드 입력이나 마우스 움직임을 처리할때
큐 사용

### 큐의 ADT
- 연산
    - boolean isFull()
    : 큐에 들어있는 데이터 개수가 maxsize인지 확인해 boolean값 반환
    - boolean isEmpty()
    : 큐에 들어있는 데이터가 하나도 없는지 확인해 boolean값 반환
    - void push()
    : 큐에 데이터 푸쉬
    - void pop()
    : 큐에서 처음에 푸시한 데이터를 팝하고, 데이터 반환
- 상태
    - int front
    : 큐에서 다음에 제거할 원소의 위치를 기록
    - int rear
    : 큐에서 최근에 푸시한 데이터의 위치를 기록
    - ItemType data[maxsize]
    : 큐의 데이터를 관리하는 배열. 최대 maxsize개의 데이터 관리



## 큐 구현하기

- 리스트 활용방식
    - 푸시연산 : append()
    - 팝 연산 : pop()
    - 팝 연산 사용 시 pop(0) 메서드에 인수로 0을 넣어줘야함
    : 맨앞의 원소가 삭제되어야하기 때문

- 덱을 큐처럼 활용하는 방식
    - 덱 DEQ (Double Ended Queue)
    - 양 끝에서 삽입 삭제 가능
    - 큐 구현 시 덱을 사용하는것이 좋다

```python
# 리스트 활용 방식 Queue
queue = []

# 큐에 데이터 추가
queue.append(1)
queue.append(2)
queue.append(3)

# 큐의 맨 앞 데이터 제거
first_item = queue.pop(0)
print(first_item) # 출력 1

# 큐에 데이터 추가
queue.append(4)
queue.append(5)

# 큐의 맨앞 데이터 제거
first_item = queue.pop(0)
print(first_item)   # 2출력
```

```python
# 덱을 큐처럼 활용하기
from collections import deque

queue = deque()

# 큐에 데이터추가
queue.append(1)
queue.append(2)
queue.append(3)

# 큐의 맨 앞 데이터 제거
first_item = queue.popleft()
print(first_item)   # 1

# 큐에 데이터 추가
queue.append(4)
queue.append(5)

# 큐의 맨 앞 데이터 제거
first_item = queue.popleft()
print(first_item)   # 2
```
